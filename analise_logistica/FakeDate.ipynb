```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Função para gerar datas aleatórias
def gerar_data_envio():
    return datetime(2025, 1, 1) + timedelta(days=random.randint(0, 100))

# Função para gerar a data de entrega com base na data de envio e uma diferença aleatória
def gerar_data_entrega(data_envio):
    return data_envio + timedelta(days=random.randint(1, 10))

# Função para gerar o modal de transporte aleatório
def gerar_modal():
    return random.choice(['Aéreo', 'Rodoviário', 'Marítimo', 'Ferroviário'])

# Gerando dados fictícios
num_pedidos = 100  # Número de pedidos no dataset
data_envios = [gerar_data_envio() for _ in range(num_pedidos)]
data_entregas = [gerar_data_entrega(data) for data in data_envios]
custos_envio = [random.uniform(5, 100) for _ in range(num_pedidos)]  # Custo de envio entre 5 e 100
pesos_produto = [random.uniform(1, 50) for _ in range(num_pedidos)]  # Peso do produto entre 1 e 50 kg
distancias = [random.uniform(10, 1000) for _ in range(num_pedidos)]  # Distância entre 10 e 1000 km
status_entrega = [random.choice(['Entregue', 'Em trânsito', 'Atrasado']) for _ in range(num_pedidos)]
modais = [gerar_modal() for _ in range(num_pedidos)]  # Modal de transporte

# Criando o DataFrame
df_logistica = pd.DataFrame({
    'ID do Pedido': range(1, num_pedidos + 1),
    'Data de Envio': data_envios,
    'Data de Entrega': data_entregas,
    'Custo de Envio': custos_envio,
    'Peso do Produto (kg)': pesos_produto,
    'Distância Percorrida (km)': distancias,
    'Status da Entrega': status_entrega,
    'Modal de Transporte': modais
})

# Salvando o DataFrame em um arquivo CSV
df_logistica.to_csv('logistica.csv', index=False)

# Exibindo as primeiras linhas do dataset
df_logistica.head()
```
