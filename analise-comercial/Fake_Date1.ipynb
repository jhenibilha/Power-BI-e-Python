```python
# Importação dos outros pacotes
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
```

```python
# Inicializa o Faker
fake = Faker()
```

```python
# Função para criar dados fictícios de uma loja com as novas colunas
def create_commercial_data_extended(num_records):
    data = []
    
    for _ in range(num_records):
        # Gerar dados fictícios
        company_name = fake.company()
        product_name = fake.word().capitalize() + " " + fake.word().capitalize()
        price = round(random.uniform(10, 500), 2)
        quantity = random.randint(1, 100)
        total_value = round(price * quantity, 2)
        date_of_sale = fake.date_this_year()
        
        # Agora, as escolhas aleatórias para cada linha
        category = random.choice(['Eletrônicos', 'Moda', 'Alimentos', 'Acessórios'])  # Categoria do produto
        region = random.choice(['São Paulo', 'Rio de Janeiro', 'Nordeste', 'Sul'])  # Região (agora aleatória por linha)
        payment_method = random.choice(['Cartão de Crédito', 'Boleto', 'Pix', 'Dinheiro'])  # Método de pagamento
        discount = random.choice([0, 5, 10, 20])  # Desconto
        sales_representative = fake.name()  # Nome do representante de vendas (usando Faker)
        sales_channel = random.choice(['Online', 'Loja Física', 'Telemarketing'])  # Canal de vendas
        promotion = random.choice(['Black Friday', 'Liquidação de Verão', 'Promoção de Natal', 'Nenhuma'])  # Promoção
        product_category = random.choice(['Eletrônicos', 'Roupas', 'Casa e Decoração', 'Saúde e Beleza'])  # Categoria do produto
        
        # Adiciona as informações à lista
        data.append([company_name, product_name, price, quantity, total_value, date_of_sale, category, region,
                     payment_method, discount, sales_representative, sales_channel, promotion, product_category])

    # Cria o DataFrame com os dados
    df = pd.DataFrame(data, columns=['Company Name', 'Product Name', 'Price', 'Quantity', 'Total Value', 'Date of Sale',
                                     'Category', 'Region', 'Payment Method', 'Discount', 'Sales Representative', 
                                     'Sales Channel', 'Promotion', 'Product Category'])

    # Converte a coluna 'Date of Sale' para o formato datetime (caso não esteja já)
    df['Date of Sale'] = pd.to_datetime(df['Date of Sale'])

    # Cria a nova coluna 'Month of Sale'
    df['Month of Sale'] = df['Date of Sale'].dt.month

    return df

# Criar 1000 registros fictícios com as novas colunas
df_commercial_data_extended = create_commercial_data_extended(1000)

# Exibir as primeiras linhas do DataFrame
df_commercial_data_extended.head()
```

```python
# Salvar o DataFrame em um arquivo CSV
df_commercial_data_extended.to_csv('dados_comerciais.csv', index=False)
```
