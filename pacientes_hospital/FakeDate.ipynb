```python
import pandas as pd
import numpy as np
from random import choice, randint
from datetime import datetime, timedelta

# Definindo parâmetros para o dataset
sexos = ['Masculino', 'Feminino']
diagnosticos = ['Pneumonia', 'Infecção Urinária', 'Fratura', 'Doenças Cardiovasculares', 'COVID-19', 'Diabetes']
num_pacientes = 1000  # Número de registros no dataset

# Gerando dados aleatórios
dados = []
for _ in range(num_pacientes):
    sexo = choice(sexos)
    idade = randint(18, 85)  # Idade entre 18 e 85 anos
    diagnostico = choice(diagnosticos)
    data_admissao = datetime.today() - timedelta(days=randint(1, 365))  # Data de admissão aleatória no último ano
    tempo_internacao = randint(1, 15)  # Tempo de internação entre 1 e 15 dias
    custo_tratamento = round(randint(1000, 5000) + np.random.rand(), 2)  # Custo aleatório entre 1000 e 5000

    dados.append([sexo, idade, diagnostico, data_admissao.strftime('%Y-%m-%d'), tempo_internacao, custo_tratamento])

# Criando o DataFrame
df = pd.DataFrame(dados, columns=['Sexo', 'Idade', 'Diagnóstico', 'Data de Admissão', 'Tempo de Internação (dias)', 'Custo do Tratamento'])

# Salvando em um arquivo CSV
df.to_csv('dados_pacientes.csv', index=False)

# Exibindo as primeiras linhas para verificação
df.head()
```
