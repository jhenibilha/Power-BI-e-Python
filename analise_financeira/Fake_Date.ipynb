```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Função para gerar uma data aleatória entre duas datas
def random_date(start_date, end_date):
    return start_date + timedelta(days=random.randint(0, (end_date - start_date).days))

# Definindo os parâmetros
clientes_por_mes = {
    1: 150,  # Janeiro
    2: 146,  # Fevereiro
    3: 152,  # Março
    4: 153,  # Abril
    5: 157,  # Maio
    6: 155   # Junho
}

formas_pagamento = ['Boleto', 'Cartão de Crédito', 'Pix']
situacoes = ['recebido', 'em aberto']
dias_vencimento = [5, 10, 20, 30]

# Função para ajustar a data de vencimento ao mês e garantir que seja válida
def ajustar_data_vencimento(mes, dia_vencimento):
    # Calcular o último dia do mês
    if mes == 2:  # Fevereiro (considerando 2024 como ano bissexto)
        ultimo_dia = 29
    elif mes in [4, 6, 9, 11]:  # Meses com 30 dias
        ultimo_dia = 30
    else:  # Meses com 31 dias
        ultimo_dia = 31

    # Ajustar a data de vencimento se o dia não for válido para o mês
    if dia_vencimento > ultimo_dia:
        dia_vencimento = ultimo_dia
    return datetime(2024, mes, dia_vencimento)

# Gerando o dataset fictício
dados = []

for mes, num_clientes in clientes_por_mes.items():
    for i in range(num_clientes):  # Número de clientes específicos por mês
        cliente = f"Cliente_{mes}_{i+1}"
        
        # Escolhendo uma data de vencimento aleatória a partir das opções de dias
        dia_vencimento = random.choice(dias_vencimento)
        data_vencimento = ajustar_data_vencimento(mes, dia_vencimento)
        
        # Decidindo se o cliente será inadimplente
        inadimplente = random.choice([True, False])  # 50% chance de inadimplente
        
        # Se o cliente for inadimplente, a data de recebimento será None (não pagou)
        if inadimplente:
            data_recebimento = None
            situacao = 'em aberto'
        else:
            # Data de pagamento, considerando que pode ser antes, no mesmo dia ou depois
            pagamento_no_dia = random.choice([True, False, False])  # 1/3 chance de pagar no dia
            if pagamento_no_dia:
                data_recebimento = data_vencimento
            else:
                # Definindo pagamento antes ou após
                if random.choice([True, False]):  # Pago antes
                    data_recebimento = random_date(data_vencimento - timedelta(days=10), data_vencimento)
                else:  # Pago após
                    data_recebimento = random_date(data_vencimento, data_vencimento + timedelta(days=15))
            
            situacao = 'recebido'
        
        # Forma de pagamento
        forma_pagamento = random.choice(formas_pagamento)
        
        # Valor aleatório entre 100 e 1000
        valor = round(random.uniform(100, 1000), 2)
        
        # Mês contrato
        mes_contrato = f"Contrato {mes}/2024"
        
        dados.append([data_vencimento.strftime('%d-%m-%Y'), 
                      data_recebimento.strftime('%d-%m-%Y') if data_recebimento else None,
                      cliente, 
                      valor, 
                      situacao, 
                      forma_pagamento, 
                      mes_contrato])

# Criando o DataFrame
df = pd.DataFrame(dados, columns=['data_vencimento', 'data_recebimento', 'nome_cliente', 'valor', 'situacao', 'forma_pagamento', 'mes_contrato'])

# Exibindo as primeiras linhas
df.head(120)
```

```python
# Salvar o DataFrame em um arquivo CSV
df.to_csv('dados_financeiros.csv', index=False)
```
